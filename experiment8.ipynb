### 1. Add few names, one name in each row, in &quot;name.txt&quot; file.
<ol type = "a">

2. The file is opened in read mode, and all names are read into a list after removing newline characters using `strip()`.  

3. The total number of names is calculated using the `len()` function.  

4. To count names starting with a vowel, a string of vowels is defined.

### 2. Store integers in a file:

<ol type = "a">
    <li>Find the maximum number</li>
    <li>Find average of all numbers</li>
    <li>Count number of numbers greater than 100</li>
</ol>

### Code: 

```python
# Creating 'numbers.txt' and storing some integers
data = [5, 120, 8, 350, 42, 15, 200]

with open("numbers.txt", "w") as file:
    for num in data:
        file.write(str(num) + "\n")

# Reading the file
with open("numbers.txt", "r") as file:
    nums = [int(line.strip()) for line in file if line.strip()]

# Finding the maximum number
max_val = max(nums) if nums else None

# Finding average
average = sum(nums) / len(nums) if nums else 0

# Counting numbers greater than 100
count_gt_100 = sum(1 for n in nums if n > 100)

# Output
print("Max Number:", max_val)
print("Average:", round(average, 2))
print("Count > 100:", count_gt_100)
```

### Explanation

1. A list of integers is created and written to a file named "numbers.txt", with each number on a new line.  

2. The file is opened in read mode, and each line is read and converted into an integer using `int()`.  

3. The maximum number is found using the `max()` function.  
   - A condition is used to avoid errors if the list is empty.  

4. The average is calculated by dividing the sum of all numbers by the total count.  

5. A generator expression is used to count how many numbers are greater than 100.  

6. Finally, the results are printed.

### 3. Assume a file city.txt with details of 5 cities.
<p>Open file city.txt and read to:</p>
<ol>
    <li>Display details of all cities</li>
    <li>Display city names with population more than 10 Lakhs</li>
    <li>Display sum of areas of all cities</li>
</ol>


```python
# Creating 'city.txt' 
city_data = [
    "Dehradun 5.78 308.20",
    "Delhi 190.0 1484.0",
    "Mumbai 184.0 603.4",
    "Nainital 0.41 11.73",
    "Bangalore 84.0 709.0"
]

with open("city.txt", "w") as file:
    for line in city_data:
        file.write(line + "\n")

# Reading the file
total_area = 0
high_pop_cities = []

print("--- All City Details ---")

with open("city.txt", "r") as file:
    for line in file:
        if line.strip():
            parts = line.split()
            
            if len(parts) == 3:   # safety check
                name = parts[0]
                pop = float(parts[1])
                area = float(parts[2])

                print(f"City: {name:10} | Population: {pop:>6}L | Area: {area:>8} sq KM")

                if pop > 10:
                    high_pop_cities.append(name)

                total_area += area

print("\n--- Cities with Population > 10 Lakhs ---")
print(", ".join(high_pop_cities))

print("\n--- Sum of Areas ---")
print(f"Total Area: {total_area:.2f} sq KM")
```

### Explanation

1. A file named "city.txt" is created and city details are written into it.  
   - Each line contains city name, population (in lakhs), and area.  

2. The file is opened in read mode to process the data.  

3. Each line is read and split into three parts: name, population, and area.  

4. The population and area values are converted into float for calculations.  

5. All city details are displayed in a formatted manner.  

6. Cities with population greater than 10 lakhs are identified and stored in a list.  

7. The total area is calculated by adding areas of all cities.  

8. Finally, the list of cities with high population and the total area are printed.

### 4. Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.
### Code:

```python
try:
    n = int(input("Enter number of test cases: "))
except ValueError as e:
    print(f"Error Code: {e}")
    n = 0

for _ in range(n):
    try:
        line = input().split()

        if len(line) < 2:
            raise ValueError("Insufficient values")

        a = int(line[0])
        b = int(line[1])

        print(a // b)

    except ZeroDivisionError as e:
        print(f"Error Code: {e}")
    except ValueError as e:
        print(f"Error Code: {e}")
```

### Explanation

1. The program first takes the number of test cases `n` as input.  
   - Exception handling is used to catch invalid input using `ValueError`.  

2. A loop runs `n` times to process each test case.  

3. In each iteration, two values `a` and `b` are taken as input in a single line and split into a list.  

4. A check is performed to ensure that two values are provided.  

5. The values are converted into integers using `int()`.  

6. Integer division (`a // b`) is performed and the result is printed.  

7. Exception handling is used to manage errors:  
   - `ZeroDivisionError`: when division by zero occurs  
   - `ValueError`: when input is not a valid integer or insufficient values  

8. If any error occurs, an appropriate error message is displayed.

### 5. Create multiple suitable exceptions for a file handling.</h2>
### Code: 

```python
class FileError(Exception):
    pass

class FileEmptyError(FileError):
    pass

class InvalidFormatError(FileError):
    pass

import os

def process_city_data(filename):
    try:
        if not os.path.exists(filename):
            raise FileNotFoundError(f"The file '{filename}' was not found.")
        
        if os.path.getsize(filename) == 0:
            raise FileEmptyError(f"'{filename}' is empty and cannot be processed.")
        
        with open(filename, 'r') as file:
            for line in file:
                if not line.strip():   # skip empty lines
                    continue

                parts = line.split()

                if len(parts) < 3:
                    raise InvalidFormatError(f"Line '{line.strip()}' is missing data.")

                print(f"Processing: {parts[0]}")

    except FileNotFoundError as e:
        print(f"System Error: {e}")
    except FileEmptyError as e:
        print(f"Data Error: {e}")
    except InvalidFormatError as e:
        print(f"Format Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        print("File operation attempt finished.")
```

### Explanation

1. Custom exception classes are created:  
   - `FileError` is the base class  
   - `FileEmptyError` is raised when the file is empty  
   - `InvalidFormatError` is raised when file content is incorrect  

2. The `os` module is used to check file existence and size.  

3. The function `process_city_data()` is defined to process file data.  

4. The program first checks if the file exists using `os.path.exists()`.  
   - If not, a `FileNotFoundError` is raised.  

5. It then checks if the file is empty using `os.path.getsize()`.  
   - If empty, a `FileEmptyError` is raised.  

6. The file is opened and read line by line.  

7. Each line is split into parts and validated.  
   - If data is missing, an `InvalidFormatError` is raised.  

8. Different exceptions are handled using `try-except` blocks with appropriate messages.  

9. The `finally` block ensures a message is printed after execution, regardless of errors.
