### 1. Add few names, one name in each row, in &quot;name.txt&quot; file.
<ol type = "a">
    <li>Count number of names</li>
    <li>Count all names starting with vowel</li>
    <li>Find longest name</li>
</ol>

```python
# Creating the file and adding the names
names_list = ["Akshit", "Vaibhav", "Deevyanshi", "Vikas", "Sna"]

with open("name.txt", "w") as file:
    for name in names_list:
        file.write(f"{name}\n")

# Reading the file to perform the tasks
with open("name.txt", "r") as file:
    names = [line.strip() for line in file if line.strip()]

# Counting number of names
total_count = len(names)

# Counting names starting with a vowel
vowels = "AEIOUaeiou"
vowel_count = sum(1 for name in names if name[0] in vowels)

# Finding the longest name
longest = max(names, key=len)

print(f"Total names: {total_count}")
print(f"Starts with vowel: {vowel_count}")
print(f"Longest name: {longest}")
```

### 2. Store integers in a file:

<ol type = "a">
    <li>Find the maximum number</li>
    <li>Find average of all numbers</li>
    <li>Count number of numbers greater than 100</li>
</ol>

### Code: 

```python
# Creating 'numbers.txt' and storing some integers
data = [5, 120, 8, 350, 42, 15, 200]

with open("numbers.txt", "w") as file:
    for num in data:
        file.write(f"{num}\n")

# Reading the file and converting strings back to integers
with open("numbers.txt", "r") as file:
    nums = [int(line.strip()) for line in file if line.strip()]

# Finding the maximum number
max_val = max(nums)

# Finding average of all numbers
average = sum(nums) / len(nums) if nums else 0

# Counting numbers greater than 100
count_gt_100 = sum(1 for n in nums if n > 100)

print(f"Max Number: {max_val}")
print(f"Average: {average:.2f}")
print(f"Count > 100: {count_gt_100}")
```

### 3. Assume a file city.txt with details of 5 cities.
<p>Open file city.txt and read to:</p>
<ol>
    <li>Display details of all cities</li>
    <li>Display city names with population more than 10 Lakhs</li>
    <li>Display sum of areas of all cities</li>
</ol>


```python
# Creating 'city.txt' 
city_data = [
    "Dehradun 5.78 308.20",
    "Delhi 190.0 1484.0",
    "Mumbai 184.0 603.4",
    "Nainital 0.41 11.73",
    "Bangalore 84.0 709.0"
]

with open("city.txt", "w") as file:
    for line in city_data:
        file.write(line + "\n")

# Reading the file and performing tasks
total_area = 0
high_pop_cities = []

print("--- All City Details ---")
with open("city.txt", "r") as file:
    for line in file:
        if line.strip():
            # Split line into [name, population, area]
            parts = line.split()
            name = parts[0]
            pop = float(parts[1])
            area = float(parts[2])
            
            # Displaying details
            print(f"City: {name:10} | Population: {pop:>6}L | Area: {area:>8} sq KM")
            
            # Checking population > 10 Lakhs
            if pop > 10:
                high_pop_cities.append(name)
            
            # Adding to total area
            total_area += area

print("\n--- Cities with Population > 10 Lakhs ---")
print(", ".join(high_pop_cities))

print("\n--- Sum of Areas ---")
print(f"Total Area: {total_area:.2f} sq KM")
```

### 4. Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.
### Code:

```python
# getting the number of test cases
try:
    n = int(input("Enter number of test cases: "))
except ValueError as e:
    print(f"Error Code: {e}")
    n = 0

for _ in range(n):
    try:
        # Reading a and b, then split by space
        line = input().split()
        
        # Ensuring that at least two values to unpack
        a = int(line[0])
        b = int(line[1])
        
        # Performing integer division
        print(a // b)
        
    except ZeroDivisionError as e:
        print(f"Error Code: {e}")
    except ValueError as e:
        print(f"Error Code: {e}")
    except IndexError:
        print("Error Code: Please provide two values (a and b)")
```

### 5. Create multiple suitable exceptions for a file handling.</h2>
### Code: 

```python
class FileError(Exception):
    # Base class for exceptions in this module
    pass

class FileEmptyError(FileError):
    # Raised when the file exists but has no content
    pass

class InvalidFormatError(FileError):
    # Raised when the file content does not match the expected format
    pass


import os

def process_city_data(filename):
    try:
        # Checking if file exists first
        if not os.path.exists(filename):
            raise FileNotFoundError(f"The file '{filename}' was not found.")
        
        # Checking if file is empty
        if os.path.getsize(filename) == 0:
            raise FileEmptyError(f"'{filename}' is empty and cannot be processed.")
        
        with open(filename, 'r') as file:
            for line in file:
                parts = line.split()
                if len(parts) < 3:
                    raise InvalidFormatError(f"Line '{line.strip()}' is missing data.")
                # Processing data
                print(f"Processing: {parts[0]}")

    except FileNotFoundError as e:
        print(f"System Error: {e}")
    except FileEmptyError as e:
        print(f"Data Error: {e}")
    except InvalidFormatError as e:
        print(f"Format Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        print("File operation attempt finished.")
```
