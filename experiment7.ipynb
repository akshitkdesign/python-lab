### 1. Write a Python function to find the maximum and minimum numbers from a sequence of numbers.</h2>
### Code:

```python
def minimumMaximum(numbers):
    if not numbers:
        return None, None
    
    maximumValue = numbers[0]
    minimumValue = numbers[0]

    for num in numbers[1:]:
        if num > maximumValue:
            maximumValue = num
        elif num < minimumValue:
            minimumValue = num

    return maximumValue, minimumValue

nums = [12, 9, 25, 15]
maximumNumber, minimumNumber = minimumMaximum(nums)
print(f"Maximum: {maximumNumber}, Minimum: {minimumNumber}")
```

### 2. Write a python function that takes a positive interger and returns the sum of the cube of all the positive integers smaller than the specified number.
### Code: 

```python
def sumOfCube(n):
    if n <= 1:
        return 0
    
    totalSum = 0
    for i in range(1, n):
        totalSum += i ** 3

    return totalSum

number1 = 12
result1 = sumOfCube(number1)

print(f"The sum of the cube of all positive integers smaller than {number1} is {result1}")

number2 = 9
result2 = sumOfCube(number2)
print(f"The sum of the cubes of all positive integers smaller than {number2} is {result2}")
```

### 3. Write a python function to print 1 to n using recursion.
### Code: 

```python
def printNumbers(n, current = 1):
    if current > n:
        return
    
    print(current)

    printNumbers(n, current + 1)

printNumbers(5)
```

### 4. Write a Python function to print 1 to n using recursion.
### Code: 

```python
print(0)
print(1)
count = 2

def fibonacci(previous1, previous2):
    global count
    if count <= 19:
        newFibonacci = previous1 + previous2
        print(newFibonacci)
        previous2 = previous1
        previous1 = newFibonacci
        count += 1
        fibonacci(previous1, previous2)
    else:
        return
    
fibonacci(1, 0)
```

### 5. Write a lambda function to find volume of cone.
### Code: 

```python
import math

def lambdaFunction(radius, height):
    volume = (1/3) * math.pi * (radius ** 2) * height
    return volume

r = 5
h = 10
print(f"The volume is {lambdaFunction(r, h)}")
```
