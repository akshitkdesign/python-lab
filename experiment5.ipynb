### 1. Write a program to count and display the number of capital letters in a given string.

```python
s = input("Enter string: ")
count = 0

for ch in s:
    if ch.isupper():
        count += 1

print("Capital letters:", count)
```

isupper() checks whether a character is uppercase.
Loop traverses the string and counts uppercase letters.

***
### 2.	Count total number of vowels in a given string.

```python
s = input("Enter string: ")
count = 0

for ch in s.lower():
    if ch in "aeiou":
        count += 1

print("Number of vowels:", count)
```

The string is converted to lowercase to handle both cases.
Each character is checked if it is a vowel.

### 3.	Input a sentence and print words in separate lines.

```python
sentence = input("Enter sentence: ")

words = sentence.split()

for word in words:
    print(word)
```

split() divides the sentence into words.
Loop prints each word on a new line.

### 4.	WAP to enter a string and a substring. You have to print the number of times that the substring occurs in the given string. String traversal will take place from left to right, not from right to left. 
* Sample Input 

* ABCDCDC 
* CDC 

* Sample Output 
* 2 


```python
string = input("Enter string: ")
substring = input("Enter substring: ")

count = 0

for i in range(len(string) - len(substring) + 1):
    if string[i:i+len(substring)] == substring:
        count += 1

print("Occurrences:", count)
```

* The program checks each possible position in the string.
* If substring matches, count is increased.

### 5.	Given a string containing both upper and lower-case alphabets. Write a Python program to count the number of occurrences of each alphabet (case insensitive) and display the same. 
* Sample Input 
* ABaBCbGc 
* Sample Output 
* 2A 
* 3B 
* 2C 
* 1G 


```python
s = input("Enter string: ").upper()

freq = {}

for ch in s:
    if ch.isalpha():
        freq[ch] = freq.get(ch, 0) + 1

for key in sorted(freq):
    print(freq[key], key)
```

* String is converted to uppercase for case-insensitive counting.
* Dictionary stores frequency of each alphabet.

### 6.	Program to count the number of unique words in a given sentence using sets. 

```python
sentence = input("Enter sentence: ")

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

* Set automatically removes duplicate words.
* Length of set gives number of unique words.

### 7.	Create 2 sets s1 and s2 of n fruits each by taking input from user and find: 
#### a.	Fruits which are in both sets s1 and s2 
#### b.	Fruits only in s1 but not in s2 
#### c.	Count of all fruits from s1 and s2 


```python
n = int(input("Enter number of fruits: "))

s1 = set()
s2 = set()

print("Enter fruits for set 1:")
for i in range(n):
    s1.add(input())

print("Enter fruits for set 2:")
for i in range(n):
    s2.add(input())

print("Common fruits:", s1 & s2)
print("Only in s1:", s1 - s2)
print("Total fruits count:", len(s1 | s2))
```

& → intersection (common items)  
- → difference (only in s1)  
| → union (all unique items)

### 8.	Take two sets and apply various set operations on them: 
S1 = {Red, yellow, orange, blue} 
S2 = {violet, blue, purple}


```python
S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Union:", S1 | S2)
print("Intersection:", S1 & S2)
print("Difference (S1 - S2):", S1 - S2)
print("Difference (S2 - S1):", S2 - S1)
```

Union → all elements  
Intersection → common elements  
Difference → elements in one set but not the other
