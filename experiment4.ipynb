### (Q1) Find a factorial of given number. 

```python
num = int(input("Enter number: "))
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial:", fact)
```

Factorial of n = n × (n-1) × ... × 1

Loop runs from 1 to n and multiplies each number.

### 2. Find whether the given number is Armstrong number. 

```python
num = int(input("Enter number: "))
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if sum == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

An Armstrong number is equal to the sum of cubes of its digits.

Example: 153 = 1³ + 5³ + 3³

### 3. Print Fibonacci series up to given term. 

```python
n = int(input("Enter number of terms: "))

a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

Each number is the sum of previous two numbers.

Sequence: 0, 1, 1, 2, 3, 5...

### 4. Write a program to find if given number is prime number or not. 

```python
num = int(input("Enter number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime Number")
else:
    print("Not Prime")
```

A prime number has only 2 factors: 1 and itself.

Loop checks divisibility from 2 to n-1.

### 5. Check whether given number is palindrome or not. 

```python
num = int(input("Enter number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

if rev == num:
    print("Palindrome")
else:
    print("Not Palindrome")
```

A palindrome reads the same forward and backward.

Example: 121, 1331

### 6. Write a program to print sum of digits. 

```python
num = int(input("Enter number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num //= 10

print("Sum of digits:", sum)
```

Extract digits using % and divide number using // to reduce it.

### 7. Count and print all numbers divisible by 5 or 7 between 1 to 100. 

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nCount:", count)
```

Checks numbers from 1 to 100 and prints those divisible by 5 or 7.

### 9. Print all prime numbers between 1 and 100. 

```python
for num in range(2, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
```

Checks each number from 2 to 100 for prime condition using loop.
