## Interactive Mode

Interactive mode allows us to execute Python commands one by one and get instant output.

It is useful for:
- Testing small code snippets
- Learning and debugging

In Jupyter Notebook, we are working in Interactive Mode.

```python
print("Hello from Interactive Mode")

x = 5
y = 10
print("Sum:", x + y)
```

Output is generated immediately after executing each cell, which shows interactive behavior.

***
## Scripting Mode

In scripting mode, Python code is written in a file (with .py extension) and executed all at once.

Steps:
1. Write code in a file (example.py)
2. Save the file
3. Run using:
   python example.py

Used for:
- Large programs
- Projects

***
### Example of Script File (example.py)

```python
# This represents a script file

print("This is a script example")

a = 20
b = 30
print("Sum:", a + b)
```

In scripting mode, all lines are executed together when the file is run.

***
## Difference Between Interactive and Scripting Mode

| Interactive Mode | Scripting Mode |
|----------------|---------------|
| Executes line by line | Executes whole program |
| Immediate output | Output after execution |
| Used for testing | Used for full programs |
| Example: Jupyter Notebook | Example: .py file |
