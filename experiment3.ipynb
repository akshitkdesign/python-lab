### 1.	Check whether a given number is divisible by 3 and 5 both. 

```python
num = int(input("Enter number: "))

if num % 3 == 0 and num % 5 == 0:
    print("Divisible by both 3 and 5")
else:
    print("Not divisible by both")
```

The program checks divisibility using modulus (%).
If remainder is 0 for both 3 and 5, the number is divisible.

### 2.	Check whether a given number is multiple of five or not. 

```python
num = int(input("Enter number: "))

if num % 5 == 0:
    print("Multiple of 5")
else:
    print("Not a multiple of 5")
```

A number is a multiple of 5 if it leaves remainder 0 when divided by 5.

### 3.	Find the greatest among two numbers. If numbers are equal than print “numbers are equal”. 

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Greatest:", a)
elif b > a:
    print("Greatest:", b)
else:
    print("Numbers are equal")
```

The program compares two numbers using if-elif-else conditions.

### 4.	Find the greatest among three numbers assuming no two values are same.  

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Greatest:", a)
elif b > c:
    print("Greatest:", b)
else:
    print("Greatest:", c)
```

The program compares all three numbers and prints the largest.

### 5.	Check whether the quadratic equation has real roots or imaginary roots. Display the roots. 

```python
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b**2 - 4*a*c

if d > 0:
    root1 = (-b + d**0.5) / (2*a)
    root2 = (-b - d**0.5) / (2*a)
    print("Real and distinct roots:", root1, root2)
elif d == 0:
    root = -b / (2*a)
    print("Real and equal roots:", root)
else:
    print("Imaginary roots")
```

Discriminant (D = b² - 4ac) determines nature of roots:
D > 0 → real & distinct  
D = 0 → real & equal  
D < 0 → imaginary

### 6.	Find whether a given year is a leap year or not. 

```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

Leap year conditions:
- Divisible by 4 and not by 100
- OR divisible by 400

### 7.	Write a program which takes any date as input and display next date of the calendar e.g. 
I/P: day=20 month=9 year=2005  
O/P: day=21 month=9 year 2005 


```python
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

# Days in months
days_in_month = [31, 28, 31, 30, 31, 30, 
                 31, 31, 30, 31, 30, 31]

# Leap year check
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    days_in_month[1] = 29

day += 1

if day > days_in_month[month - 1]:
    day = 1
    month += 1

if month > 12:
    month = 1
    year += 1

print("Next Date:", day, "/", month, "/", year)
```

The program:
1. Stores days in each month
2. Adjusts February for leap year
3. Increments day
4. Handles month/year change if limit exceeds

### 8.	Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and calculate the percentage. 
##### CGPA=percentage/10 
##### CGPA range: 
##### 0 to 3.4 -> F 
##### 3.5 to 5.0->C+ 
##### 5.1 to 6->B 
##### 6.1 to 7-> B+ 
##### 7.1 to 8-> A 
##### 8.1 to 9->A+ 
##### 9.1 to 10-> O (Outstanding) 
 
##### Sample Gradesheet 
##### Name: Rohit Sharma 
##### Roll Number: R17234512 SAPID: 50005673 
##### Sem: 1 Course: B.Tech. CSE AI&ML 
 
##### Subject name: Marks 
##### PDS: 70 
##### Python: 80 
##### Chemistry: 90 
##### English: 60 
##### Physics: 50 
##### Percentage: 70% 
##### CGPA:7.0 
##### Grade:  A

```python
name = input("Enter name: ")
roll = input("Enter roll number: ")
sapid = input("Enter SAP ID: ")

m1 = int(input("PDS marks: "))
m2 = int(input("Python marks: "))
m3 = int(input("Chemistry marks: "))
m4 = int(input("English marks: "))
m5 = int(input("Physics marks: "))

total = m1 + m2 + m3 + m4 + m5
percentage = total / 5
cgpa = percentage / 10

# Grade calculation
if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O"

print("\n--- Grade Sheet ---")
print("Name:", name)
print("Roll No:", roll)
print("SAP ID:", sapid)

print("\nMarks:")
print("PDS:", m1)
print("Python:", m2)
print("Chemistry:", m3)
print("English:", m4)
print("Physics:", m5)

print("\nPercentage:", percentage)
print("CGPA:", cgpa)
print("Grade:", grade)
```

The program:
1. Takes marks of 5 subjects
2. Calculates percentage = total / 5
3. Calculates CGPA = percentage / 10
4. Assigns grade based on CGPA range using if-elif
5. Displays a formatted grade sheet
