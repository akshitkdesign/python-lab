### 1.	Scan n values in range 0-3 and print the number of times each value has occurred. 

```python
n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    val = int(input("Enter value (0-3): "))
    lst.append(val)

for i in range(4):
    print(i, "occurs", lst.count(i), "times")
```

### Explanation

* The program takes the number of elements `n` as input.  

* An empty list `lst` is created to store the values.  

* Using a loop, `n` values are entered by the user and added to the list.  

* A loop runs from 0 to 3 to check each required number.  

* The `count()` method is used to find how many times each number appears in the list.  

* The result is printed for each value.

### 2.	Create a tuple to store n numeric values and find the average of all values. 

```python
n = int(input("Enter number of elements: "))
values = []

for i in range(n):
    values.append(float(input("Enter value: ")))

t = tuple(values)

avg = sum(t) / len(t)

print("Tuple:", t)
print("Average:", avg)
```

### Explanation

* The program takes `n` numeric values as input from the user.  

* These values are stored in a list first.  

* The list is then converted into a tuple using the `tuple()` function.  

* The average is calculated by dividing the sum of all elements by the total number of elements.  

* Finally, the tuple and its average are displayed.

### 3.	WAP to input a list of scores for N students in a list data type. Find the score of the runner-up and print the output. 
* Sample Input 
* N = 5 
* Scores= 2 3 6 6 5 
* Sample output 
* 5 
* Note: Given list is [2, 3, 6, 6, 5]. The maximum score is 6, second maximum is 5. Hence, we print 5 as the runner-up score.

```python
n = int(input("Enter number of students: "))
scores = list(map(int, input("Enter scores: ").split()))

scores = list(set(scores))   # remove duplicates
scores.sort()

print("Runner-up score:", scores[-2])
```

### Explanation

* The program takes the number of students `n` and their scores as input.  

* The scores are stored in a list.  

* Duplicate values are removed by converting the list into a set.  

* The unique values are converted back to a list and sorted in descending order.  

* The second element in the sorted list represents the runner-up score.  

* This value is printed as the output.

### 4.	Create a dictionary of n persons where key is name and value is city.  

#### a.	Display all names 
#### b.	Display all city names 
#### c.	Display student names and cities of all students. 
#### d.	Count the number of students in each city. 


```python
n = int(input("Enter number of students: "))
students = {}

for i in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    students[name] = city

print("Names:", list(students.keys()))
print("Cities:", list(students.values()))

for name, city in students.items():
    print(name, "->", city)

city_count = {}
for city in students.values():
    city_count[city] = city_count.get(city, 0) + 1

print("Students per city:", city_count)
```

### Explanation

* The program takes the number of students `n` as input.  

* A dictionary is created where the key is the student's name and the value is the city.  

* Names and cities are entered by the user and stored in the dictionary.  

* All student names are displayed using `keys()` method.  

* All cities are displayed using `values()` method. 

* Name-city pairs are displayed using `items()` method.  

* A new dictionary is used to count how many students belong to each city.  

* The count of students per city is printed.

### 5.	Store details of n movies in a dictionary by taking input from the user. Each movie must store details like name,  year, director name, production cost, collection made (earning) & perform the following :- 

```python
n = int(input("Enter number of movies: "))
movies = []

for i in range(n):
    name = input("Name: ")
    year = int(input("Year: "))
    director = input("Director: ")
    cost = float(input("Production Cost: "))
    collection = float(input("Collection: "))

    movie = {
        "name": name,
        "year": year,
        "director": director,
        "cost": cost,
        "collection": collection
    }

    movies.append(movie)

print("\nAll Movie Details:")
for m in movies:
    print(m)

print("\nMovies before 2015:")
for m in movies:
    if m["year"] < 2015:
        print(m["name"])

print("\nProfitable Movies:")
for m in movies:
    if m["collection"] > m["cost"]:
        print(m["name"])

search_director = input("\nEnter director name: ")
for m in movies:
    if m["director"].lower() == search_director.lower():
        print(m["name"])
```

### Explanation

* The program takes the number of movies `n` as input.  

* For each movie, details like name, year, director, cost, and collection are entered.  

* These details are stored in a dictionary and added to a list of movies.  

* All movie details are printed.  

* Movies released before 2015 are filtered and displayed.  

* Movies that made a profit (collection > cost) are identified and printed. 

* The user is asked to enter a director's name.  

* Movies directed by that specific director are displayed.
