### 1.	Scan n values in range 0-3 and print the number of times each value has occurred. 

```python
n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    val = int(input("Enter value (0-3): "))
    lst.append(val)

for i in range(4):
    print(i, "occurs", lst.count(i), "times")
```

* The program stores values in a list.
* count() function counts how many times each number (0–3) appears.

### 2.	Create a tuple to store n numeric values and find the average of all values. 

```python
n = int(input("Enter number of elements: "))
values = []

for i in range(n):
    values.append(float(input("Enter value: ")))

t = tuple(values)

avg = sum(t) / len(t)

print("Tuple:", t)
print("Average:", avg)
```

* Values are stored in a tuple.
* Average is calculated using sum() and len().

### 3.	WAP to input a list of scores for N students in a list data type. Find the score of the runner-up and print the output. 
* Sample Input 
* N = 5 
* Scores= 2 3 6 6 5 
* Sample output 
* 5 
* Note: Given list is [2, 3, 6, 6, 5]. The maximum score is 6, second maximum is 5. Hence, we print 5 as the runner-up score.

```python
n = int(input("Enter number of students: "))
scores = list(map(int, input("Enter scores: ").split()))

scores = list(set(scores))   # remove duplicates
scores.sort()

print("Runner-up score:", scores[-2])
```

* Duplicate scores are removed using set().
* List is sorted and second highest value is printed.

### 4.	Write a recursive function to print Fibonacci series upto n terms. 

```python
n = int(input("Enter number of persons: "))
d = {}

for i in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    d[name] = city

# a) Names
print("Names:", d.keys())

# b) Cities
print("Cities:", d.values())

# c) Name + City
for name, city in d.items():
    print(name, "lives in", city)

# d) Count students in each city
city_count = {}

for city in d.values():
    city_count[city] = city_count.get(city, 0) + 1

print("Students per city:", city_count)
```

* Dictionary stores name as key and city as value.
* keys(), values(), items() are used for access.
* Another dictionary counts number of students per city.

### 5.	Write a lambda function to find volume of cone. 

```python
n = int(input("Enter number of movies: "))
movies = []

for i in range(n):
    print("\nEnter details for movie", i+1)
    name = input("Name: ")
    year = int(input("Year: "))
    director = input("Director: ")
    cost = float(input("Production Cost: "))
    collection = float(input("Collection: "))

    movie = {
        "name": name,
        "year": year,
        "director": director,
        "cost": cost,
        "collection": collection
    }

    movies.append(movie)

# a) Print all movie details
print("\nAll Movies:")
for m in movies:
    print(m)

# b) Movies before 2015
print("\nMovies before 2015:")
for m in movies:
    if m["year"] < 2015:
        print(m["name"])

# c) Movies with profit
print("\nProfitable Movies:")
for m in movies:
    if m["collection"] > m["cost"]:
        print(m["name"])

# d) Movies by specific director
search_dir = input("\nEnter director name: ")
print("Movies by", search_dir)

for m in movies:
    if m["director"].lower() == search_dir.lower():
        print(m["name"])
```

Each movie is stored as a dictionary inside a list.

- List stores multiple movies
- Dictionary stores movie details

Operations performed:
a) Display all movies  
b) Filter movies before 2015  
c) Check profit (collection > cost)  
d) Search movies by director
